I started by importing `networkx` since I am going to be using that to visualize the network.

```python
import networkx as nx
```

Next I initiated a networkx graph object that we will be building in.

```python
g = nx.Graph()
```

I then created a list of edges for the network and added them to the graph object.

```python
edge_lst = [
    ('Andre','Carol'),('Carol', 'Fernando'),('Fernando','Andre'),('Carol','Diane'),
    ('Andre','Diane'),('Andre','Beverly'),('Beverly','Diane'),('Beverly','Garth'),
    ('Beverly', 'Ed'),('Ed','Diane'),('Ed','Garth'),('Garth','Fernando'),
    ('Diane','Fernando'),('Diane','Ed'),('Diane','Garth'),
    ('Heather','Fernando'),('Heather','Garth'),('Heather','Ike'),('Ike','Jane')
    ]
g.add_edges_from(edge_lst)
```

I couldn't get the graph to look good on its own so I decided to manually set the node locations to get it to look nice. Same with the labels.

```python
pos = {
    'Ed': (1,0),
    'Beverly': (0,0.5),
    'Garth': (2,0.5),
    'Diane': (1,1),
    'Andre': (0,1.5),
    'Fernando': (2,1.5),
    'Carol': (1,2),
    'Heather': (3,1),
    'Ike': (4,1),
    'Jane': (5,1)
}
pos_lab = {
    'Ed': (1,-0.15),
    'Beverly': (0,0.35),
    'Garth': (2,0.35),
    'Diane': (1,0.85),
    'Andre': (0,1.35),
    'Fernando': (2,1.35),
    'Carol': (1,1.85),
    'Heather': (3,.85),
    'Ike': (4,.85),
    'Jane': (5,.85)
}
```

Finally I visualized the graph and the labels for the nodes.

```python
nx.draw(g, pos)
nx.draw_networkx_labels(g, pos_lab)
```
